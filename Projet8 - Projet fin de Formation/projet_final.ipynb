# Contexte de l'étude

La qualité du système éducatif est souvent présentée comme un facteur clé du développement économique. C'est dans ce cadre que SACMEQ (Southern and Eastern Africa Consortium for Monitoring Educational Quality) a mis en place des politiques éducatives, des reformes ainsi des programmes éducatifs permettant gérer les défis liés à la qualité de l’enseignement en Afrique australe et orientale (voir figure 1). Comparativemnt aux autres pays d'Afrique subsaharienne, les pays composant le consortium SACMEQ ont enregistré des efforts remarquables en termes de scolarisations.
 
Malgre les avancés, l'Afrique australe et orientale est l'une des régions où les inégalités scolaires sont plus importantes. Nous pouvons citer les inégalités régionales, entre les zones rurales et urbaines, entre les isolées et non isolées, entre les filles et garçons, entre etablissements publics et privés etc.

Partant de ces constats, l'ojectif de ce travail est de prédire la réussite scolaire de l'élève compte tenu des caractéristiques socioéconomique de milieu. Au passage, un accent sera mis sur la distance qui sépare le domicile de l'élève et son école en prenant en compte les moyens d'accès à lécole.

### Categorical transformation 

```python
#XQPERMNT : 
africa['XQPERMNT'] = africa['XQPERMNT'].replace(['PERMANENT GOVT', 'PERMANET NOT-GOVT'], 'PERMANENT')
africa['XQPERMNT'] = africa['XQPERMNT'].replace(['TEMPORARY GOVT', 'TEMPORARY NON-GOVT'], 'TEMPORARY')

# zpmealsc : Free school meals
africa['zpmealsc'] = africa['zpmealsc'].replace(['Two or more per day',"One per day"], 'Yes')

# PMOTHER : Mother's Education
africa['PMOTHER'] = africa['PMOTHER'].replace(['No School, No Adult Education',
                                               'No School, Some Adult Education'], 'No School Level')
africa['PMOTHER'] = africa['PMOTHER'].replace(['All Primary', 'Completed Some Primary', 'Completed All Primary',
                                               'Some Training After Primary'], 'Primary Level')
africa['PMOTHER'] = africa['PMOTHER'].replace(['Completed Some Secondary', 'Completed All Secondary',
                                               'Completed Training After Secondary'], 'Secondary Level')
africa['PMOTHER'] = africa['PMOTHER'].replace(['Completed University Degree', 
                                               'Completed Some University'], 'University Level')

# PFATHER : Father's Education
africa['PFATHER'] = africa['PFATHER'].replace(['No School, No Adult Education',
                                               'No School, Some Adult Education'], 'No School Level')
africa['PFATHER'] = africa['PFATHER'].replace(['All Primary', 'Completed Some Primary', 'Completed All Primary',
                                               'Some Training After Primary'], 'Primary Level')
africa['PFATHER'] = africa['PFATHER'].replace(['Completed Some Secondary', 'Completed All Secondary',
                                               'Completed Training After Secondary'], 'Secondary Level')
africa['PFATHER'] = africa['PFATHER'].replace(['Completed University Degree', 
                                               'Completed Some University'], 'University Level')

# XQPROFES : Teacher Training
africa['XQPROFES'] = africa['XQPROFES'].replace(['2 YRS TT', '3 YRS TT', '>3 YRS TT'], 'TWO OR MORE YEARS')
africa['XQPROFES'] = africa['XQPROFES'].replace(['1 YR TT', '<1 YR TT' ], 'ONE YEAR LESS')
africa['XQPROFES'] = africa['XQPROFES'].replace(['NO TT' ], 'NO TEACHER TRAINING')

# PNURSERY : Pupil Preschool
africa['PNURSERY'] = africa['PNURSERY'].replace(['ONE YEAR', 'TWO YEARS', 'THREE OR MORE YEARS',
                                                 'A FEW MONTHS'],'ONCE OR MORE')

# PENGLISH : Pupil speaks English At Home
africa['PENGLISH'] = africa['PENGLISH'].replace(['MOST OF THE TIME'], 'ALL THE TIME')

# PLIGHT : Source Of Lighting
africa['PLIGHT'] = africa['PLIGHT'].replace(['PARAFFIN/OIL', 'CANDLE', 'FIRE', 'GAS'], 'NO LIGHTING')

# PREPEAT : Repeated Grades
africa['PREPEAT'] = africa['PREPEAT'].replace(['ONCE', 'TWICE', 'THREE PLUS'], 'ONCE OR MORE')

# SLOCAT : School Location
africa['SLOCAT'] = africa['SLOCAT'].replace(['LARGE CITY', 'SMALLTOWN'], 'URBAN')
africa['SLOCAT'] = africa['SLOCAT'].replace(['ISOLATED'], 'RURAL')
```

## Kmeans

```python
X = africa[['lon', 'lat']]
```

```python
X.info()
```

```python
sc= StandardScaler()
X_scaled = pd.DataFrame(sc.fit_transform(X.values), columns=X.columns)
```

```python
distortions = []
K = range(1,30)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X_scaled)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(9,6))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('WC55')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

We can observe that the “elbow” is the number 8 which is optimal for this case. Now we can run a K-Means using as n_clusters the number 8.

#### - Quelles sont les méthodes à essayer ?
Parce que nous essayons de prédire une valeur, nous allons examiner les méthodes de régression.
```python
africa = africa.drop(['schoolname','zmalocp','zralocp', 'province','district', 'lon', 'lat'], axis =1)
```

```python
africa = africa[['country','zpses' ,'clusters','PSEX', 'PNURSERY', 'PENGLISH', 'PTRAVEL2','PMOTHER',
                'PFATHER', 'PLIGHT', 'PREPEAT', 'STYPE', 'SLOCAT','XSEX', 'XQPERMNT', 'XQPROFES', 
                 'zpmealsc', 'zphmwkhl', 'SPUPPR04', 'SPUPPR13', 'SPUPPR14','DISTANCE','PABSENT',
           'XNUMYRS','numstu','zpsibs','PAGE','zraloct', 'zmaloct','PSCORE']]
```

#### Separate target variable Y from features X

```python
print("Separating labels from features...")
features_list = ['country', 'zraloct', 'zmaloct', 'PAGE', 'DISTANCE', 'XNUMYRS', 'numstu', 'PSEX',
       'PNURSERY', 'PENGLISH', 'PTRAVEL2', 'PMOTHER', 'PFATHER', 'PLIGHT','PABSENT', 'PREPEAT', 'STYPE', 'SLOCAT', 
                 'XSEX', 'XQPERMNT', 'XQPROFES', 'zpsibs', 'zpmealsc', 'zphmwkhl', 'zpses', 'SPUPPR04', 'SPUPPR13',
       'SPUPPR14', 'clusters']
target_variable = 'PSCORE'

Z = africa.loc[:,features_list]
Y = africa.loc[:,target_variable]

print("...Done.")
print()
```

```python
Z = pd.get_dummies(Z, drop_first = True) ## Toujours mettre drop_first = True pour éviter le piège des dummies
Z.head()
```

#### Features normalisation

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(Z)
```

#### Divide dataset Train set & Test set 

```python
from sklearn.model_selection import train_test_split
print("Dividing into train and test sets...")
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
print("...Done.")
print()
```

## Random Forest Regressor

Enfin, après le prétraitement, nous pouvons maintenant commencer à prédire avec un RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 100,
                              criterion='mse',
                              max_depth = 30,
                              n_jobs = -1,
                              max_features=0.2,
                              oob_score = True,
                              bootstrap = True,
                              min_samples_leaf = 3)
model.fit(X_train, Y_train)

print('Score de Train : {:.4f} \n Test Score: {:.4f} \n OOB Score: {:.4f}'.format(model.score(X_train, Y_train), 
                                                                                 model.score(X_test, Y_test),
                                                                                 model.oob_score_))
```

#### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
rf_reg_grid = RandomForestRegressor()
params = {"n_estimators": [40, 100], 'max_depth': range(5, 30, 5),
          'min_samples_leaf': range(2, 5, 1)}
regressor = GridSearchCV(cv=5, estimator = rf_reg_grid, param_grid = params).fit(X_train, Y_train)
regressor.best_params_
```

#### Predictions on training set

```python
print("Predictions on training set...")
Y_train_pred = model.predict(X_train)
print("...Done.")
print(Y_train_pred)
print()
```

#### Predictions on test set

```python
print("Predictions on test set...")
Y_test_pred = model.predict(X_test)
print("...Done.")
print(Y_test_pred)
print()
```

```python
from sklearn.metrics import r2_score
print("R2 score on training set : ", r2_score(Y_train, Y_train_pred))
print("R2 score on test set : ", r2_score(Y_test, Y_test_pred))
```

# Conclusion et recommandations

Le pays de provenance joue aussi un rôle important dans la variation du résultat scolaire de l'élève. Toutefois, nous pouvons signalé quelques nuances dans notre analyse. Par exemple, 90% des scores dans certains tels que le Swaziland, le Kenya et la Tanzanie sont supérieurs à la moyenne alors que les pays comme l'Afrique du Sud, le Malawi et la Zambie, ont 90% des scores inférieurs à 480. Celà nous amène à faire l'hypothèse que certains pays ont probablement sur-évalué les scores de leurs mauvais élèves et que d'autres ont sous-évalué les scores de leurs bons élèves. Celà peut être lié à l'hétérogénéité des politiques éducatives qui diffèrent d'un pays à l'autre. Ce qui peut être une perspective de nouvelle analyse que nous n'avons pas abordé dans ce présent projet.

- la mise en place des bus scolaires dans certaines zones,
