## Description du challenge

Ceci est un projet de data visualisation et d'exploration de données. 
Notre but va être d'essayer de comprendre **l'amour** !  Un sujet bien compliqué.  
- Pour simplifier, notre but ici est plutôt de comprendre ce qu'il se passe durant un speed dating et surtout de comprendre ce qui va influencer l'obtention d'une **deuxième rendez-vous**.
- Notre objectif principal est de comprendre chacune des variables présentes dans le dataset. 

- Nous allons donc faire une analyse descriptive des principaux facteurs qui influencent le fait d'obtenir un deuxième rendez-vous.
- Les données ont été recueillies auprès de **552 participants** à des speed dating expérimentaux entre 2002 et 2004.
Au cours de ces événements, les participants avaient un "premier rendez-vous" de quatre minutes avec tous les autres participants du sexe opposé.
A la fin de ces quatre minutes, les participants étaient invités à dire s'ils souhaitaient revoir leur partenaire ou non. Il leur était également demandé d'évaluer leur partenaire en fonction de six caractéristiques :(**Attractivité, Sincérité, Intelligence, Amusant, Ambition et Intérêts partagés**).
- L'ensemble de données comprend également les données des questionnaires recueillis auprès des participants à différents moments du processus. Ces champs comprennent :(les données démographiques, les habitudes de rencontre, la perception de soi à travers les attributs clés, les convictions sur ce que les autres trouvent précieux chez un compagnon, les informations sur le mode de vie).

(Voir le site **Speed Dating Data Key (1).doc** pour le dictionnaire de données et la clé des questions).

# II. Data Visualization & Data Analyse

## Age distribution by gender

```python
Speed_Dating['gender'] = Speed_Dating.gender.map({1 : 'Male', 0 : 'Female'}).fillna(Speed_Dating.gender)
Speed_Dating.gender.value_counts(dropna=False)
```

- Dans notre dataset, il y a un peu plus de mâles que de femelles.
- On peut vérifier si cette différence peut aussi être vue dans la différence d'âge ..

```python
g = sns.FacetGrid(Speed_Dating, hue='gender', height = 6)
g.map(plt.hist, 'age', alpha= 0.7, bins=20)
g.set_xlabels('Age', fontsize=12)
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Age distribution by gender', fontsize=15)
g.add_legend()
```

```python
Speed_Dating[['gender', 'age']].groupby('gender').agg(['mean', 'median', 'min', 'max', 'count'])
```

```python
ax = Speed_Dating[['gender', 'age']].groupby('gender').mean().plot(kind='bar', figsize=(12,5), legend=False, 
                                               title='Mean Age by Gender',
                                                ylim=(23,27))
ax.set_xticklabels(['Female', 'Male'], fontsize=12, rotation='horizontal')

ax.set_xlabel('', fontsize=1)

for i in ax.patches:
    ax.text(i.get_x()+.20, i.get_height()+.05, \
            str(round((i.get_height()), 1)), fontsize=15)
```

- Il n'y a pas de grandes différences dans la moyenne d'âge, mais nous pouvons voir que les hommes sont, en moyenne, légèrement plus âgés que les femmes. Cependant, la différence au niveau de l'âge médian peut être dû à la présence d'un certains nombres d'outliers **(voir le Boxplot ci-apès)**

## Boxplot (outliers visualisation)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='age', data=Speed_Dating, hue='gender', showmeans = True, meanline=True)
```

## La repartition des participants selon leur race 

```python
Speed_Dating['race'] = Speed_Dating.race.map({1: 'Black', 2: 'White', 3: 'Hispanic', 
                          4: 'Asian', 6: 'Other'}).fillna(Speed_Dating.race)

Speed_Dating.race.value_counts(dropna=False)
```

- Participants sont majoritairement caucasiens, puis asiatiques. Cela pourrait être soit une affaire culturelle, soit que ces événements aient été organisés dans des zones spécifiques avec cette démographie.

```python
Speed_Dating[['race', 'age']].groupby('race').agg(['mean', 'median', 'min', 'max', 'count'])
```

```python
ax = Speed_Dating[['race', 'age']].groupby('race').mean().plot(kind='bar', figsize=(12,5), legend=False,
                                               title='Mean Age by Race',
                                                   ylim=(24,28), color='rgbmy')
ax.set_xticklabels(['Asian', 'Black', 'Hispanic',  'Other', 'White'], 
                   fontsize=12, rotation='horizontal')
ax.set_xlabel('',fontsize=1)

for i in ax.patches:
    ax.text(i.get_x()+.155, i.get_height()+.05, \
            str(round((i.get_height()), 1)), fontsize=12)
```

```python
Speed_Dating_Race = Speed_Dating.groupby(['iid','race', "gender"]).size().reset_index()
print(Speed_Dating_Race.race.value_counts().sum())
Speed_Dating_Race['gender']=Speed_Dating_Race.gender.map({1 : 'Male', 0 : 'Female'})
Speed_Dating_Race['race'] = Speed_Dating_Race.race.map({1 : 'Black/African', 2 : 'European/Caucasian', 
                                                        3 : 'Latino/Hispanic', 4 : 'Asian/Pacific',
                                                       5 : 'Native', 6 : 'Other'})

sns.catplot(x='race', hue='gender', data= Speed_Dating_Race, kind ="count", aspect = 1.4)
plt.show()
```

## Proportion des participants ayant obtenus un deuxième rendez-vous

```python
#Création des catégories yes et no:
match_reponse = Speed_Dating.match.apply(lambda x: "Yes" if x==1
                                      else "No" )
match_numb = match_reponse.value_counts()
#print(match_numb)

plt.figure(figsize=(8,6))
plt.title("Proportion des participants ayant obtenus un deuxième rendez-vous")
plt.pie(match_numb, labels=match_numb.index, autopct='%1.1f%%', explode=[0.0,0.2],
        shadow=True, startangle=35, radius=1.2)

plt.show()
```

```python
Speed_Dating['field_cd'] = Speed_Dating.field_cd.map({1: 'Law', 2: 'Math', 3: 'Soc. Sc.', 4: 'Med. Sc.',
                                 5: 'Eng.', 6: 'Journ.', 7: 'Hist.', 8: 'Econ', 9: 'Educ.',
                                 10: 'Nat. Sc.', 11: 'Soc. Wr.', 12: 'Und.', 13: 'Pol. Sc.',
                                 14: 'Film', 15: 'Arts', 16:'Lang.', 17: 'Arch.', 18: 'Oth.'}).fillna(Speed_Dating.field_cd)

#Speed_Dating.field_cd.value_counts(dropna=False)
```

## Career Female vs Male

```python
tmp = Speed_Dating[['gender', 'field_cd']].groupby(['field_cd', 'gender']).size().unstack().fillna(0)
ax = tmp.plot(kind='bar', figsize=(12,6), stacked=True)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=12, rotation=45)

ax.set_title('Field of study by gender', fontsize=18)
ax.set_xlabel('',fontsize=1)

plt.show()
```
